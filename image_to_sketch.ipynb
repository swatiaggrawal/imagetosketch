```python
#import necessary libraries
import cv2
from google.colab.patches import cv2_imshow
```

```python
img=cv2.imread("/marvel-iron-man-lede.webp") #read image
```

```python
grayimg=cv2.cvtColor(img,cv2.COLOR_BGR2GRAY) #convert to gray image
```

```python
invgrayimg=255-grayimg #invert the gray image
```

```python
blurinvgrayimg=cv2.GaussianBlur(invgrayimg,(19,19),0) # blur the inverted gray image using gaussian filter
```

```python
invblurimg=255-blurinvgrayimg #invert the blurred gray image
```

```python
sketch=cv2.divide(grayimg,invblurimg,scale=256.0) #forming sketch
```

```python
cv2_imshow(img) #original image
```

```python
cv2_imshow(sketch) #sketch of original image
```
